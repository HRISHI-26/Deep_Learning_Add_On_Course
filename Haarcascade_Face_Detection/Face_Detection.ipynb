```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the Daar Cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "sources/haarcascade_frontalface_default.xml")
```

```python
# Load an image
image = cv2.imread("sources/face_1.jpg")
```

```python
# Convert image into grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
# Perform Face Detection
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30,30))
```

```python
#Detect and draw boxes around faces
for(x,y,w,h) in faces:
  cv2.rectangle(image, (x,y), (x+w, y+h), (0, 255, 0), 3)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(8,6))
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```
