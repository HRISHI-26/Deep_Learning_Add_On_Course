##Model Creation

```python
# sklearn contain various type of models
# linear_model is one of them

from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# Just changing column names for simplicity
df.columns = ["sl","sw","pl","pw","sp"]
```

```python
x = df[["sl","sw","pl","pw"]]
y = df["sp"]
```

```python
x
```

```python
y
```

Splitting into training and testing data

```python
                                                              # to mix up training and testing data
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.3, random_state=100)
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
ytrain
```

Training

```python
lr = LogisticRegression()
```

```python
mf = lr.fit(xtrain,ytrain)
```

Prediction

```python
mf.predict(xtest)
```

```python
mf.score(xtest,ytest)
```
